```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
crimeData = pd.read_csv("crimeDataWithPhase.csv")
yearList = [2008,2009,2010,2011,2012,2013,2014,2015]
yearData = [crimeData[(crimeData.DateOccured >= (str(x)+"-01-01")) & (crimeData.DateOccured < (str(x+1)+"-01-01"))] for x in yearList]
```

```python
fullMoons = [yearData[x]["DateOccured"][yearData[x].Phase == "Full Moon"].count() for x in range(len(yearList))]
newMoons = [yearData[x]["DateOccured"][yearData[x].Phase == "New Moon"].count() for x in range(len(yearList))]
firstQuarters = [yearData[x]["DateOccured"][yearData[x].Phase == "First Quarter"].count() for x in range(len(yearList))]
lastQuarters = [yearData[x]["DateOccured"][yearData[x].Phase == "Last Quarter"].count() for x in range(len(yearList))]
```

```python
years = np.array(yearList)
plt.plot(years, fullMoons, label="Full Moon")
plt.plot(years, newMoons, label="New Moon")
plt.plot(years, firstQuarters, label="First Quarter")
plt.plot(years, lastQuarters, label="Last Quarter")
plt.ylabel("Number of Crimes")
plt.xlabel("Year")
plt.grid()
plt.title("Crimes by Lunar Phase from 2008-2015")
plt.legend()
plt.show()
```
